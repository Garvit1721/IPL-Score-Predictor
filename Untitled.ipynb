```python
y = data_2[data_2['match_id']==1]
```

```python
y[y['inning']==2]
```

```python
y[y['inning']==1]
```

```python
y.drop(['batsman','non_striker','bowler','is_super_over','wide_runs','bye_runs','legbye_runs','noball_runs','penalty_runs','batsman_runs','extra_runs','dismissal_kind','fielder'],axis=1,inplace=True)
```

```python
y['over'] = y['over']-1
```

```python
y['ball_count'] = y['over']*6+y['ball']
```

```python
y['cumulative_runs'] = y.groupby('inning')['total_runs'].cumsum()
```

```python
y.drop(['over','ball'],axis=1,inplace=True)
```

```python
y['Team1'] = y['batting_team'][0]
y['Team2'] = y['bowling_team'][0]
```

```python
y
```

```python
total_balls = 120
y['ball_left_with_team1'] = y.apply(lambda row: total_balls - row['ball_count'] if row['inning'] == 1 else None, axis=1)
y['ball_left_with_team2'] = y.apply(lambda row: total_balls - row['ball_count'] if row['inning'] == 2 else 120, axis=1)
```

```python
y['ball_left_with_team1'].fillna(0, inplace=True)
y['ball_left_with_team2'].fillna(120, inplace=True)
```

```python
y
```

```python
final_score_team1 = y[y['inning'] == 1]['cumulative_runs'].max()
final_score_team2 = y[y['inning'] == 2]['cumulative_runs'].max()
y['run_by_team1'] = y.apply(lambda row: row['cumulative_runs'] if row['inning'] == 1 else final_score_team1, axis=1)
y['run_by_team2'] = y.apply(lambda row: row['cumulative_runs'] if row['inning'] == 2 else 0, axis=1)
y
```

```python
y['runs_left'] = y['run_by_team1'] - y['run_by_team2']
```

```python
y
```

```python
y['wickets_team1'] = y.apply(lambda row: 1 if row['inning'] == 1 and pd.notna(row['player_dismissed']) else 0, axis=1)
y['wickets_team2'] = y.apply(lambda row: 1 if row['inning'] == 2 and pd.notna(row['player_dismissed']) else 0, axis=1)
y['wickets_team1'] = y.groupby('inning')['wickets_team1'].cumsum()
y['wickets_team2'] = y.groupby('inning')['wickets_team2'].cumsum()
final_wickets_team1 = y[y['inning'] == 1]['wickets_team1'].max()
y['wickets_team1'] = y.apply(lambda row: final_wickets_team1 if row['inning'] == 2 else row['wickets_team1'], axis=1)
y
```

```python
y['over'] = (y['ball_count'] - 1) // 6 + 1
y['runs_in_over'] = y.groupby(['inning', 'over'])['cumulative_runs'].diff().fillna(y['cumulative_runs'])
y['cumulative_runs_in_over'] = y.groupby(['inning', 'over'])['runs_in_over'].cumsum()
y['current_run_rate'] = y['cumulative_runs_in_over'] / y['over']
over_wise_rr = y.drop_duplicates(subset=['inning', 'over'], keep='first')[['inning', 'over', 'current_run_rate']]
```

```python
y.drop(['runs_in_over','cumulative_runs_in_over'],axis=1,inplace=True)
```

```python
y
```

```python
y['required_runs'] = y.apply(lambda row: max(0, row['run_by_team1'] - row['run_by_team2']) if row['inning'] == 2 else 0, axis=1)
y['remaining_overs'] = y.apply(lambda row: 20 - (row['over'] - 1) if row['inning'] == 2 else 0, axis=1)
y['required_run_rate'] = y['required_runs'] / y['remaining_overs']
y
```

```python
y.drop(['over','required_runs','remaining_overs'],inplace=True,axis=1)
```

```python
y.drop(['batting_team','bowling_team','total_runs','player_dismissed','ball_count','cumulative_runs'],axis=1,inplace=True)
```

```python
y['required_run_rate'].fillna(y['current_run_rate'],inplace=True)
y
```
